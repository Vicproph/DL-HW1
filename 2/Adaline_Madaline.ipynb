# Adaline

```python

class Adaline:
    def __init__(self, inputs, targets):
        self.inputs = inputs
        self.targets = targets
        np.random.seed(42)
        self._weights = np.random.random(self.inputs.shape[1])
        self._bias = np.random.random()
    
    def fit(self, lr0.001, thr=0.0001, max_epochs=1000):
        self.lr = lr
        self.thr = thr
        self.losses = []

        for epoch in range(max_epochs):
            errors = []
            
            for i in range(self.inputs.shape[0]):
                y = np.dot(self.inputs[i], self._weights) + self._bias
                error = 0.5 * ((self.targets[i] - y) ** 2)
                update = self.lr * (self.targets[i] - y) 
                self._weights += update * self.inputs[i]
                self._bias += update
                errors.append(error)

            loss = np.mean(errors)
            print(f'Epoch {epoch+1}, Loss = {loss}')
            self.losses.append(loss)
            
            if loss < self.thr:
                print(f'Converged after {epoch+1} epochs.')
                break

        return self._weights, self._bias, self.losses
    
    def predict(self, test_data):
        predictions = []
        for i in range(test_data.shape[0]):
            y = np.dot(test_data[i], self._weights) + self._bias
            predictions.append(y)
        return np.array(predictions)
```

```python
model = Adaline(x_train.values, y_train.values)
_,_, losses = model.fit(0.01, 0.015)
```

```python
plt.plot(losses)
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
X_moons, Y_moons = make_moons(n_samples=500, noise=0.1, random_state=42)
X_train_moons, X_test_moons, Y_train_moons, Y_test_moons = train_test_split(X_moons, Y_moons, test_size=0.5, random_state=3)
plt.scatter(X_moons[Y_moons == 0][:, 0], X_moons[Y_moons == 0][:, 1], color='red', label='Class 0')
plt.scatter(X_moons[Y_moons == 1][:, 0], X_moons[Y_moons == 1][:, 1], color='blue', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title("Moon-shaped Dataset")
plt.show()
```

```python
from sklearn.metrics import accuracy_score

model = Adaline(X_train_moons, Y_train_moons)
_,_, losses = model.fit(0.1, 0.01 ,max_epochs=40)

pred = model.predict(X_test_moons)
pred = [1 if p >= 0.5 else 0 for p in pred]
accuracy = accuracy_score(Y_test_moons, pred)

print("Adaline Accuracy:", accuracy)
```

```python
import numpy as np

class Madaline:
    def __init__(self, input_size, hidden_size, output_size, lr=0.2, thr=0.001):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.lr = lr
        self.thr = thr

        self.hidden_weights = np.random.random((input_size, hidden_size))
        self.hidden_bias = np.random.random(hidden_size)
        self.output_weights = np.random.random((hidden_size, output_size))
        self.output_bias = np.random.random(output_size)

    def fit(self, inputs, targets, max_epochs=1000):
        for epoch in range(max_epochs):
            hidden_outputs = np.dot(inputs, self.hidden_weights) + self.hidden_bias
            output = np.dot(hidden_outputs, self.output_weights) + self.output_bias

            for i in range(inputs.shape[0]):
                hidden_error = targets[i] - hidden_outputs[i]
                output_error = targets[i] - output[i]

                for j in range(self.input_size):
                    for k in range(self.hidden_size):
                        self.hidden_weights[j][k] += self.lr * hidden_error[k] * inputs[i][j]
                    self.hidden_bias[k] += self.lr * hidden_error[k]

                for j in range(self.hidden_size):
                    for k in range(self.output_size):
                        self.output_weights[j][k] += self.lr * output_error[k] * hidden_outputs[i][j]
                    self.output_bias[k] += self.lr * output_error[k]

            loss = 0.5 * np.mean((targets - output) ** 2)
            print(f'Epoch {epoch+1}, Loss = {loss}')
            if loss < self.thr:
                print(f'Converged after {epoch+1} epochs.')
                break

    def predict(self, test_data):
        hidden_outputs = np.dot(test_data, self.hidden_weights) + self.hidden_bias
        output = np.dot(hidden_outputs, self.output_weights) + self.output_bias
        return output
```

```python
np.random.seed(42)
input_size = 2
hidden_size = 3
output_size = 1

madaline = Madaline(input_size, hidden_size, output_size, lr=0.001)
madaline.fit(X_train_moons, Y_train_moons)

pred = madaline.predict(X_test_moons)
pred = [1 if p >= 0.5 else 0 for p in pred]
accuracy = accuracy_score(Y_test_moons, pred)
print("4 madaline Accuracy:", accuracy)
```
